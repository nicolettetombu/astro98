```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1

np.random.seed(0)
x = np.random.randint(0, 101, size = 40)
x2 = np.arange(len(x))

np.random.seed(1)
y = np.random.randint(0, 101, size = 40)
y2 = np.arange(len(y))

np.random.seed(2)
z = np.random.randint(0, 101, size = 40)
z2 = np.arange(len(z))
```

```python
fig, axs = plt.subplots(2,1)
axs[0].plot(x2, x, color = 'orange', linewidth = '10' )
axs[0].plot(y2, y, color = 'red', linestyle = 'dashed' )
plt.xlabel('position')
plt.ylabel('value')
plt.show


axs[1].scatter(z2,z, marker = 'D', color = 'm' )
plt.xlabel('position')
plt.ylabel('value')
plt.show
```

```python
#2

R = 0.083144
a = 16.02
b = 0.1124
temp = []
for P in range(1,11):
    for V in range(10,31):
        T = (((P + 1)/(V**2))*(V - b))/R
        temp.append(T)             

temp = np.array(temp)
temp = temp.reshape((10, 21))
x = (np.arange(10, 31), np.arange(1, 11))

plt.contourf(np.arange(10, 31), np.arange(1, 11), temp, vmin = 0, vmax = 30)
plt.xlabel('volume')
plt.ylabel('pressure')
plt.colorbar()
plt.show

```
