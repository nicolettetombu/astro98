```python
import pandas as pd

data = pd.read_csv("satellites.csv")
data
```

```python
magnitude_list = data['magnitude'].tolist()

print(magnitude_list)

#converting data to a list to clean
```

```python
magnitude_numbers = []

for value in magnitude_list:
    parts = value.split('')
    magnitude_numbers.append(parts[0])
    
print(magnitude_numbers)

#cleaning data, removing variables and symbols
```

```python
clean_magnitude_list = []

for value in magnitude_numbers:
    parts = value.split('R')
    clean_magnitude_list.append(parts[0])

print(clean_magnitude_list)
```

```python
clean2_magnitude_list = []

for value in clean_magnitude_list:
    parts = value.split('V')
    clean2_magnitude_list.append(parts[0])
    
print(clean2_magnitude_list)
```

```python
clean3_magnitude_list = []

for value in clean2_magnitude_list:
    parts = value.split('r')
    clean3_magnitude_list.append(parts[0])
    
print(clean3_magnitude_list)
```

```python
final_magnitude_list = [x for x in clean3_magnitude_list if '?' not in x]

print(final_magnitude_list)
```

```python
#cleaning radius data, first converting to list

radius_list = data['radius'].tolist()

print(radius_list)
```

```python
#removing +- symbol 

clean_radius_list = []

for value in radius_list:
    parts = value.split('')
    clean_radius_list.append(parts[0])
    
print(clean_radius_list)
```

```python
#removing the values that correspond to the question marks in magnitude list to keep values aligned to same celestial body and to keep lists same length
final_radius_list = []

for i, item in enumerate(clean3_magnitude_list):
    if item != '?':
        final_radius_list.append(clean_radius_list[i])
        
print(final_radius_list)
```

```python
import matplotlib.pyplot as plt
import numpy as np

R = np.array(final_radius_list, dtype = float)
M1 = np.array(final_magnitude_list, dtype = float)
d = np.pi*R*((10**(M1+26.74))**(1/5))

np.random.seed(3)
M2 = np.random.randint(0,22, size = 100)
R2 = 133
d2 = np.pi*R2*((10**(M2+26.74))**(1/5))

plt.figure()
plt.scatter(M1,d, marker = '.', color = 'red', label = 'actual data')
plt.scatter(M2, d2, marker = '.', color = 'blue', label = 'model equation')


plt.xlabel('magnitude')
plt.ylabel('distance')
plt.legend(loc = 'upper left')
plt.title('Relationship between Magnitude and Distance of Celestial Bodies')
plt.show()
```
