```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = np.sin(x)
```

```python
plt.figure()
plt.plot(x,y)
plt.title('sine')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x2 = np.linspace(0, 2*np.pi, 100)
y2 = x2**2
```

```python
plt.figure(figsize=(7.3,17))
plt.plot(x2,y2, color="pink", marker="2", 
        linestyle="dashdot")
plt.title("x squared")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,1].plot(x,y2)
axs[1,0].plot(x2,y)
axs[1,1].plot(x2,y2)
plt.show()
```
